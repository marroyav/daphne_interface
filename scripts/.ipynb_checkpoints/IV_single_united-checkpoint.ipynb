This script allows you to analyzed a SINGLE IV CURVE (with bias) step by step, just by specifing the file path.
ATTENTION: the input file can be whatever.

```python
import uproot
import numpy as np
import pandas as pd
from numpy.polynomial import Polynomial
from scipy.optimize import curve_fit
from scipy.signal import find_peaks
from scipy.signal import peak_widths
from scipy.signal import peak_prominences
from scipy.signal import savgol_filter
from scipy.stats import johnsonsu
from matplotlib import pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
def derivative(X, Y):
    dx = np.diff(X)
    dy = np.diff(Y)
    return dy/dx
```

INPUT:

```python
filename = '/nfs/home/anbalbon/iv_curves_pdii/data/Apr-17-2024-run02/Apr-17-2024_2256_IvCurves_trim_np04_apa2_ip10.73.137.109/apa_2_afe_0_ch_10.root'
```

Information about the IV curve: 

```python
plt_name = (filename.split('/')[-1]).split('.')[0]
ip_address = (filename.split('ip')[-1]).split('/')[0]
endpoint = ip_address.split('.')[-1]
apa = (plt_name.replace('apa_','')).split('_')[0]
ch = int(plt_name.split('ch_')[-1])
afe =  ch//8
data = (filename.split('/')[-2]).split('_')[0]
time = (filename.split('/')[-2]).split('_')[1]
SiPM_type = mapping(ip_address, ch)
```

Read the input file as dataframe - SELECT THE CORRECT VERSION:

```python
if ('mar' in data.lower()) and ('2024' in data) and (int(data.split('-')[1]) <14): 
    # OLD DATA VERSION - before 14/03/2024 - see IV_single_1 and IV_endpoint_1
    print('Old data version - before 14/03/2024')
    file = uproot.open(filename)
    bias_list = file['tree']['IV']['bias'].array(library="np")
    current = file['tree']['IV']['current'].array(library="np")
    #current = np.flip(file['tree']['IV']['current'].array(library="np"))
    df = pd.DataFrame({'V':bias_list, 'I':current})
    Vbd_bias_dac = 0
    version = 0
elif ('mar' in data.lower()) and ('2024' in data) and (int(data.split('-')[1]) >=14) and (int(data.split('-')[1]) <19): 
    # MIDDLE DATA VERSION - between 14/03/2024 and 19/03/2024 - see IV_single_2 and IV_endpoint_2
    print('Middle data version - between 14/03/2024 and 19/03/2024')
    file = uproot.open(filename)
    trim = file['tree']['IV']['trim'].array(library="np")
    #current = file['tree']['IV']['current'].array(library="np")
    current = np.flip(file['tree']['IV']['current'].array(library="np"))
    bias = file['tree']['run']['bias'].array(library="np")[0]
    start_time = file['tree']['run']['time_start'].array(library="np")[0]
    end_time = file['tree']['run']['time_end'].array(library="np")[0]
    df = pd.DataFrame({'V':trim, 'I':current})
    Vbd_bias_dac = bias[-1]
    version = 1
else : 
    # NEW DATA VERSION - from 19/03/2024 - see IV_single_3 and IV_endpoint_3
    print('New data version - from 19/03/2024')
    file = uproot.open(filename)
    trim = file['tree']['iv_trim']['trim'].array(library="np")
    #current = file['tree']['iv_trim']['current'].array(library="np") 
    #current = np.flip(file['tree']['iv_trim']['current'].array(library="np")) 
    # current = (file["tree/iv_trim/current"].array())[::-1]
    current = (file["tree/iv_trim/current"].array())[::-1] * (-1)
    bias_dac=file['tree']['bias']['bias_dac'].array(library="np")
    Vbd_bias_dac = bias_dac[-1]
    bias_V=file['tree']['bias']['bias_v'].array(library="np")
    start_time = file['tree']['run']['time_start'].array(library="np")[0]
    end_time = file['tree']['run']['time_end'].array(library="np")[0]
    df = pd.DataFrame({'V':trim, 'I':current})
    version = 2
```

Data check:

```python
for i in range(len(df) - 2):
    if all(df.iloc[i:i+3]['I'] > 0):
        df = df.iloc[i:]
        break

if (df['I'].isna().all()):
    print('ERROR: Current is all NaN')
    print(df)
elif (df['I'].isna().sum() >= 10):
    print('ERROR: More than 10 NaN value for current')
    print(df)
elif (df['I'].isna().sum() > 0) and (df['I'].isna().sum() < 10):
    print('Warning: some NaN value for current, less than 10 NaN. Let continue the analysis')

if ((df['I'] < 0).all()) :
    print('ERROR: Current is negative')
    print(df)

if (df['I'] < 100).all():
    print('ERROR: Too low current')
    print(df)
    print(df['I'].max())

if (len(df['I']) < 20):
    print('ERROR: Few points, less than 20')
    print(df)

if ((df.loc[df.index[-1], 'I'] - df.loc[0, 'I'])) < 100 : 
    print('ERROR: Wrong current range')
    #print(df)
```

Plot of the reverse IV curve + a savgol filter:

```python
import matplotlib.ticker as ticker
fig, ax = plt.subplots()
fig.suptitle('REV IV - ENDPOINT:' + str(endpoint) + ' APA:' + str(apa) + ' AFE:' + str(afe) + ' CH:' + str(ch) + ' SiPM: ' + SiPM_type)
ax.set_xlabel("Trim (DAC)")
ax.set_ylabel("Current") # unit of measure ?
#ax.set_yscale('log')
ax.scatter(df['V'], df['I'], marker='o',s=5, color='blue', label='Acquired IV curve')
ax.xaxis.set_major_locator(ticker.MultipleLocator(500))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(100))
#plt.axvline(x=2218)

# First filter on current
#savgol_window = (len(df['V'])) // 15  # 5 seems okay -> it depends on the number of points
step = round((df['V'].diff()).mean())
print('Step = ' + str(step))
if (step > 40) : 
    savgol_window = 4  # 5 seems okay -> it depends on the step 
elif (step <= 40) and (step > 10): 
    savgol_window = 5
else:
    savgol_window = 7



print('Savgol window = ' + str(savgol_window))
df['I_savgol'] = savgol_filter(df['I'],savgol_window,1) 
ax.scatter(df['V'], df['I_savgol'], marker='o',s=5, color='deepskyblue', label='Savgol filtered IV curve')
print(df['I_savgol'])
ax.legend()
```

Plot of the first normalized derivative (1/I * dI/dV) + second savgol filter on it:

```python
fig_der, ax_twin = plt.subplots()
fig_der.suptitle('REV IV - ENDPOINT:' + str(endpoint) + ' APA:' + str(apa) + ' AFE:' + str(afe) + ' CH:' + str(ch) + ' SiPM: ' + SiPM_type)
ax_twin.set_ylabel('Normalized Derivative')
df['der_norm'] = [i / j for i, j in zip(derivative(df['V'], df['I']), df['I'])] + [0] 
#df['der_norm'] = np.gradient(df['I']) /df['I']
ax_twin.scatter(df["V"],df['der_norm'], marker='o', s=5, color='purple', label='Derivative of acquired data (no filter)')

df['der_I_savgol'] = [i / j for i, j in zip(derivative(df['V'], df['I_savgol']), df['I_savgol'])] + [0] 
ax_twin.scatter(df['V'],df['der_I_savgol'], marker='o', s=5, color='fuchsia', label='Derivative of filtered data')

# Second filter 
df['der_I_savgol_2'] = savgol_filter(df['der_I_savgol'],savgol_window*3,2)

if df['der_I_savgol_2'].isna().all() :
    print('ERROR: stop and check you data')
    print(df)
else:
    ax_twin.scatter(df["V"],df['der_I_savgol_2'], marker='o', s=5, color='orange', label='Derivative of filtered data + Savgol filter')
    #ax_twin.set_ylim([df['der_I_savgol_2'].min()*0.3, df['der_I_savgol_2'].max()*2])

ax_twin.legend()
```

Search for the maximum of first derivative (filtered):

```python
# First method: find just the maximum of filtered derivativa
#peak_index = df['der_I_savgol_2'].idxmax()
#ax_twin.scatter(df.loc[peak_index, 'V'],df.loc[peak_index, 'der_I_savgol_2'], marker='o', s=25, color='yellow', label='Maximum point')

peak_index= (df.iloc[7:-7]['der_I_savgol_2']).idxmax()
print(peak_index)
while(df.loc[peak_index, 'I']<1):
    peak_index= (df.iloc[peak_index+3:-7]['der_I_savgol_2']).idxmax()
    print(peak_index)
ax_twin.scatter(df.loc[peak_index, 'V'],df.loc[peak_index, 'der_I_savgol_2'], marker='o', s=25, color='yellow', label='Maximum point')

fig_der
```

Select few data around the maximum 

```python
half_point_range = (len(df['V'])) // 12  # 6 seems okay
if (step > 40) : 
    half_point_range = 6  # 5 seems okay -> it depends on the step 
elif (step <= 40) and (step > 10): 
    half_point_range = 8
else:
    half_point_range = 10
min_index = peak_index - half_point_range
if min_index < 0:
    min_index = 0
max_index = peak_index + half_point_range 
if max_index < 0:
    max_index = len(df['V'])-1
df_fit = df.loc[min_index:max_index]

ax_twin.scatter(df_fit['V'],df_fit['der_I_savgol_2'], marker='o', s=15, color='red', label='Selected points for fit')

ax_twin.legend()
fig_der
```

Do a parabolic fit (2nd degree polynomial) of the selected data:

```python
#poly2_par, poly2_coeff = Polynomial.fit(df_fit['V'], df_fit['der_I_savgol_2'], deg = 2, full = True)  
poly2_coeff = np.polyfit(df_fit['V'], df_fit['der_I_savgol_2'], 2)

x_poly2 = np.linspace(df_fit.at[df_fit.index[0],'V'], df_fit.at[df_fit.index[-1],'V'], 100) 

#y_poly2 = [Polynomial(poly2_par.convert().coef)(i) for i in x_poly2]
y_poly2 = np.polyval(poly2_coeff, x_poly2)

ax_twin.plot(x_poly2, y_poly2, color='green', label = '2nd deg fit')

ax_twin.set_ylim([df_fit['der_I_savgol_2'].min()*0.3, df_fit['der_I_savgol_2'].max()*2])

ax_twin.legend()
fig_der
```

Search for the breakdown voltage, as the maximum of the parabola:

```python
Vbd_trim_dac = x_poly2[np.argmax(y_poly2)]
ax_twin.axvline(x=Vbd_trim_dac, color='lime' ,linestyle='--', label= r'$V_{bd} \rightarrow $ bias= '+ str(Vbd_bias_dac) + ' trim= '+f'{Vbd_trim_dac:.0f} (DAC)')
print('Breakdown voltage:')
print('Bias = ' + str(Vbd_bias_dac) + ' DAC')
print('Trim = ' + str(round(Vbd_trim_dac)) + ' DAC')
#ax_twin.set_xlim([V_bd-50,V_bd+50])
ax_twin.legend()

fig_der
```

TRIM DAC-VOLT conversion:

```python
#Vbd_trim_dac = 2000
Vbd_trim_dac = 2500 - Vbd_trim_dac
Vbd_trim_V = Vbd_trim_dac * 0.0011
print('Trim --> ' + str(round(Vbd_trim_dac)) + ' DAC = ' + str(round(Vbd_trim_V,3)) + ' V' )
```

BIAS DAC - VOLT conversion:

```python
fig_dacV, ax_dacV = plt.subplots()
fig_dacV.suptitle('Bias conversion: DAC vs VOLT - ENDPOINT:' + str(endpoint) + ' APA:' + str(apa) + ' AFE:' + str(afe) + ' CH:' + str(ch))
ax_dacV.set_ylabel("Bias (V)")
ax_dacV.set_xlabel("Bias (DAC)") 
ax_dacV.scatter(bias_dac, bias_V, marker='o',s=6, color='blue',label = 'Bias scan')

DACvsV_coeff = np.polyfit(bias_dac, bias_V, 1)
ax_dacV.plot(bias_dac, np.polyval(DACvsV_coeff, bias_dac), color='cyan', label = f'Linear fit y={DACvsV_coeff[0]:.3f}x+{DACvsV_coeff[1]:.3f} (V)')

Vbd_bias_V = round(DACvsV_coeff[0]*Vbd_bias_dac + DACvsV_coeff[1],2)
#ax_dacV.hlines(y=Vbd_bias_V, xmin=bias_dac.min(), xmax=bias_dac.max(), color='r',linestyle = 'dotted')
#ax_dacV.vlines(x=Vbd_bias_dac, ymin=bias_V.min(), ymax=bias_V.max(), color='r',linestyle = 'dotted')
print('Bias --> ' + str(Vbd_bias_dac) + ' DAC = ' + str(Vbd_bias_V) + ' V' )
print('Slope: ' + str(DACvsV_coeff[0]))
print('Intercept: ' + str(DACvsV_coeff[1]) )
ax_dacV.legend()
```

Breakdown voltage evaluation:

```python
Vbd_V = Vbd_bias_V - Vbd_trim_V
print('The breakdown voltage is: ' + str(round(Vbd_V,3)) + ' V')

if SiPM_type == 'fbk':
    Vbd_roomT = 32.5
    Vbd_ln2T = 26.8
elif SiPM_type == 'hpk':
    Vbd_roomT = 51.1
    Vbd_ln2T = 41.3
else: 
    Vbd_roomT = 0
    Vbd_ln2T = 0

print('For ' + SiPM_type + ' Vbd is ' + str(Vbd_roomT) + 'V at roomT and ' + str(Vbd_ln2T) + 'V at ln2T')
```
